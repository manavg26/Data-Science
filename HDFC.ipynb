```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
pd.set_option('display.max_columns',200)

df = pd.read_csv('HDFCBANK.csv')
df.dtypes
```

```python
pd.to_datetime(df['Date'])
df.dtypes

df.shape
```

```python
df.isna().sum()
# new_df = df.loc[~df.duplicated(subset='Open')].reset_index(drop=True).copy()

ax = df['Adj Close'].value_counts().head(50).plot(kind='bar',title='Closing price')

ax.set_xlabel('Price')
ax.set_ylabel('count')
ax

```

```python
ax = df['Adj Close'].plot(kind='hist',bins=500) #histogram
```

```python
ax = df['Adj Close'].plot(kind='kde') # kernel density
```

```python
df.plot(kind='scatter',x='Volume',y='Adj Close')
plt.show()
```

```python
sns.scatterplot(x='Volume',y='Adj Close',data=df)
```

```python
sns.pairplot(vars=['Low','Adj Close','High'],hue='Volume',data=df)
```

```python
df_corr=df[['Low','Adj Close','High','Volume']].dropna().corr() #co relation b/w columns

sns.heatmap(df_corr,annot=True)
```
